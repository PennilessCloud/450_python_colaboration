```python
from control import matlab ## you get this via this pip command: pip install control
import numpy as np
from scipy import signal as sig
import matplotlib.pyplot as plt

num = [100, 300]
den1 = [1, 2]
den2 = [1, 5]
den3 = [1, 6, 25]

d12 = sig.convolve(den1,den2)
denT = sig.convolve(d12,den3)

system = sig.lti(num,denT)

w, Hmag, Hphase = sig.bode(system, 100)

##This is still not working, so I am attempting to fix this
##gm, pm, wg, wp = matlab.margin(Hmag, Hphase, w)
##print(pm)


plt.figure()
plt.semilogx(w,Hmag)
plt.grid(which="both")
plt.figure()
plt.semilogx(w,Hphase)
plt.grid(which="both")
plt.show()
```
