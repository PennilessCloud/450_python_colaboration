```python
# Test Script for Simulation with Python

import numpy as np
import matplotlib.pyplot as plt
from scipy import signal as sig

mn = -2
mx = 3
numpts = 100

w = np.logspace(mn,mx,numpts)

num = [100, 300]
d1 = [1, 2]
d2 = [1, 5]
d3 = [1, 6, 5]

d23 = sig.convolve(d2,d3)
den = sig.convolve(d1,d23)

# Create w (frequency range) and H (the transfer function)
w, H = sig.freqs(num,den,worN=w)

# Produce the magnitude of H:
Hmag = 20 * np.log10(abs(H))

# Produce the angle of H:
# Use "unwrap" function to keep all angles from wrapping
# to erroneous positive values.
# Use "degrees" function to convert radians to degrees.
Hang = np.degrees( np.unwrap( np.angle(H) ) )

plt.plot(w, Hmag)
plt.xscale("log")
plt.grid(which="both")
plt.show()

plt.plot(w, Hang)
plt.xscale("log")
plt.grid(which="both")
plt.show()
```
